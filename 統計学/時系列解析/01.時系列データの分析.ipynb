```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# 日本語フォントを設定
font = {'family': 'IPAexGothic'}
mpl.rc('font', **font)

%matplotlib inline
```

# データセットの読み込み

```python
import pandas_datareader.data as web
import datetime

# 日経平均株価
ni225 = web.DataReader("NIKKEI225", "fred", "2000/1/30")
ni225.head()
```

```python
start = datetime.datetime(1990, 7, 2)
end = datetime.datetime(2019, 7, 2)

# 為替(USD/JPY)
dollar_yen_df = web.DataReader("DEXJPUS", "fred", start, end)
dollar_yen_df.head()
```

# データの図示
時系列解析の第一歩として、まずデータを図示してみることが重要である。これにより、時系列の大まかな特徴を捉えることができるばかりでなく、今後どのような解析を行うべきか、その方針をたてることができる。

```python
plt_title = '為替レート'
time_series = dollar_yen_df
```

```python
plt.figure(figsize=(16, 8))
plt.title(plt_title)
plt.plot(time_series)
plt.grid()
plt.show()
```

# データの定常化
時系列データに非定常性が見られる場合には、解析に進む前に、データの前処理を行うことがある。

## 変数変換

```python
plt_title = '対数変換した為替レート'
time_series = np.log(dollar_yen_df)
```

```python
plt.figure(figsize=(16, 8))
plt.title(plt_title)
plt.plot(time_series)
plt.grid()
plt.show()
```

## 差分

```python
plt_title = '差分系列'
time_series = dollar_yen_df.diff()
```

```python
plt.figure(figsize=(16, 8))
plt.title(plt_title)
plt.plot(time_series)
plt.grid()
plt.show()
```

```python
plt_title = '対数値の差分系列'
time_series = np.log(dollar_yen_df).diff()
```

```python
plt.figure(figsize=(16, 8))
plt.title(plt_title)
plt.plot(time_series)
plt.grid()
plt.show()
```
