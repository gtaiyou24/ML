```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100) + 2 #平均値が+2ずれる
```

```python
#それぞれ描画する
sns.distplot(data1)
sns.distplot(data2)
```

# ボックスプロット
ボックスプロット：データの分布を描画する。

```python
# boxplotを作ります。
sns.boxplot(data=[data1,data2])
```

真ん中の線＝中央値

上下の(箱の)線＝それぞれ25%ずつ

箱＝25% - 75%までのデータ

点＝外れ値

### 外れ値を含むボックスプロット

```python
#whis=np.infとすると外れ値を含むヒゲになる
sns.boxplot(data=[data1,data2],whis=np.inf)
```

### 水平に並べる

```python
# whis=np.inf : 外れ値を含ませる
# orient：方向という意味
# orient='h' : 横方向
# orient='v' : 縦方向
sns.boxplot(data=[data1,data2],whis=np.inf, orient = 'h')
```

## Violin/Viola plot はbox plotのシンプルさとKDEの情報量を兼ね備えます。

```python
# 正規分布に従う乱数です。
data1 = stats.norm(0,5).rvs(100)

# γ(ガンマ)分布に従う乱数を生成します。
data2 = np.concatenate([stats.gamma(5).rvs(50)-1, -1*stats.gamma(5).rvs(50)])

sns.boxplot(data=[data1,data2],whis=np.inf)
```

この描画を見る限り、データは変わらないように見えるが
実は違うということをヴァイオリンプロットで示す！

# ヴァイオリンプロット

```python
# ヴァイオリンプロットを描くと違いが分かります。
sns.violinplot(data =[data1,data2])
```

０（正規分布）：０を中央に両端に分布が広がる

1（ガンマ分布）：０が中心だが、山が２つある。

```python
# バンド幅を細かくしてみましょう。
sns.violinplot(data=data2,bw=0.01)
```

```python
# rugプロットと同じような機能もあります。
sns.violinplot(data=data1,inner="stick")
```
