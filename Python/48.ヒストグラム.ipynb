```python
from numpy.random import randn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

# ヒストグラム
<font size='3'>データ分析で未知のデータを見たときは、まずヒストグラムを描いてそのデータがどういった分布をしているのかをチェックする</p>

```python
#正規分布に従う値を100個取得
dataset1 = randn(100)
```

```python
plt.hist(dataset1)
```

 - 縦軸：　データの個数
 - 横軸：　データ幅

### 色を変える

```python
dataset2 = randn(80)
```

```python
# colorに引数を渡す
plt.hist(dataset2, color='indianred')
```

### 標準化する
それぞれのデータに全体のうち何%属しているかを縦軸に表示

```python
# normed=Trueとするだけ
plt.hist(dataset1, normed=True)
```

標準化することで2つのデータ分布を描写することができる

```python
# データビンを20に細かくする
# 重なると分からなくなるのでalpha=0.5とする
plt.hist(dataset1, normed=True, alpha=0.5, bins=20)
plt.hist(dataset2, normed=True, alpha=0.5, bins=20, color='indianred')
```

### 同時分布を描画

```python
data1 = randn(1000)
data2 = randn(1000)
```

```python
# joinplotを使う
sns.jointplot(data1, data2)
```

上図だと少しわかりにくいので2次元の図を使用する

```python
# kind='hex'とするだけ
sns.jointplot(data1, data2, kind='hex')
```
