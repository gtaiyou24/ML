```python
from numpy.random import randn
import seaborn as sns
%matplotlib inline
```

# データを可視化するSeabornの便利な関数: distplot

```python
dataset = randn(100)
```

### データの**ヒストグラム**と**カーネル密度推定**の２つを描く

```python
#dist:分布
sns.distplot(dataset)
```

### データの**カーネル密度推定**と**ラグプロット**を描く

```python
#ヒストグラムは描かないのでhist=Falseとする
sns.distplot(dataset, rug=True, hist=False)
```

distplotは幾つかのプロットが同時に描かれるのでそれぞれのグラフのスタイルを変更できる

### それぞれの色を変える

```python
sns.distplot(dataset, bins=25,
             kde_kws={'color':'indianred','label':'KDE PLOT'},
             hist_kws={'color':'blue','label':"HISTOGRAM"})
```

### Seriesから描写する

```python
from pandas import Series

ser1 = Series(dataset,name='My_DATA')
```

```python
# pandasのSeriesを描画します
sns.distplot(ser1,bins=25)
```
